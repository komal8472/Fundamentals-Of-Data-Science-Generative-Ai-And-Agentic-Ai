# Match case statement

```python
number=2

match number:
 case 1:
     print('one')
 case 2 | 3:
     print('two or three')
 case _:
     print('other number')
```

```python
number =3

match number:
  case 1:
      print('one')
  case 2:
      print('two')
  case 3 |4:
      print('three or four')
  case _:
      print('other number')
```

```python
# example- This program checks which of the three numbers is the largest. If all numbers are equal, it prints that too.

a = 7
b = 9
c = 3
if a>b and a>c:
    print(a,'is large')
elif b>a and b>a:
    print(b,' is large')
elif c>a and c>b:
    print(c,' is large')
else:
    print('both are same')
```
